```python
raw_df = spark.read.format('csv').\
option('header','true').option('mode','DROPMALFORMED')\
.load('2008.csv')
```

```python
raw_df.count()
```

```python
raw_df.printSchema()
```

```python
raw_df\
.select(['DepTime','TaxiOut','TaxiIn',\
                     'DayOfWeek','Distance','ArrDelay'])\
                     .describe().show()
```

```python
raw_df.show(20)
```

```python
raw_df.groupBy('month').count().orderBy('count',ascending=False).show()
```

```python
from pyspark.sql import functions as sparkf
```

```python
raw_df.groupBy('month').agg({'month':'count','ArrDelay':'avg'})\
.withColumnRenamed('count(month)','numMonth')\
.withColumnRenamed('avg(ArrDelay)','avgArrDelay')\
.orderBy(sparkf.col('numMonth').desc(),sparkf.col('avgArrDelay').desc()).show()
```

```python
raw_df.groupby('DepTime').count().orderBy('DepTime').show()
```

```python
def time2period(input):

  try:
    x = int(input)
    if 1 <= (x) <= 559:
      return "00.01-05.59"
    elif 600 <= (x) <= 1159:
      return "06.00-11.59"
    elif 1200 <= (x) <= 1759:
      return "12.00-17.59"
    elif 1800 <= (x) <= 2400:
      return "18.00-24.00"
    else:
      return "null"
  except ValueError:
    return "null"
  
```

```python
time2period("10.-")
```

```python
timeperiod = sparkf.udf(lambda x: time2period(x))
```

```python
raw_df.withColumn('DepTime',timeperiod(raw_df.DepTime)).groupby('DepTime').count().orderBy('DepTime').show(10)
```
